# Função triângular

## Defining the Numpy array for Tip Quality

```python
x_qual = np.arange(0, 11, 0.1)
```

## Defining the Numpy array for Triangular membership functions

```python
qual_lo = sk.trimf(x_qual, [1, 5, 9])
```

## Plot of Triangular membership functions

```python
plt.plot(x_qual, qual_lo)
plt.grid()
```

### Velocidade

```python
#Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 80, 0.1)

#Defining the triangular membership functions

slow = sk.trimf(x, [30, 30, 50])
medium = sk.trimf(x, [30, 50, 70])
medium_fast = sk.trimf(x, [50, 60, 80])
full_speed = sk.trimf(x, [60, 80, 80])

#Plotting the Membership Functions Defined
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
plt.grid()
```

# Fuzzy sets operation

## Union

\begin{equation}
\mu_{(A \cup B)} = \max(\mu_A(x),\mu_B(x)) = \mu_A(x) \wedge \mu_B(x)
\end{equation}

### Defining the Numpy array for Tip Quality

```python
x_qual = np.arange(0, 11, 0.1)
```

### Defining the Numpy array for two membership functions (Triangular)

```python
qual_lo = sk.trimf(x_qual, [0, 0, 5])
qual_md = sk.trimf(x_qual, [0, 5, 10])
```

### Finding the Maximum (Fuzzy Or)

```python
fuzzy_out = sk.fuzzy_or(x_qual, qual_lo, x_qual, qual_md)
```

### Plot

```python
plt.plot(x_qual, qual_lo)
#plt.grid()
#plt.show()

plt.plot(x_qual, qual_md)
#plt.grid()
#plt.show()

plt.plot(fuzzy_out[0], fuzzy_out[1])
plt.grid()
plt.show()

```

## Intersection

\begin{equation}
\mu_{(A \cap B)} = \min(\mu_A(x),\mu_B(x)) = \mu_A(x) \vee \mu_B(x)
\end{equation}

### Defining the Numpy array for Tip Quality

```python
# np
```

### Defining the Numpy array for two membership functions (Triangular)

```python
qual_lo = sk.trimf(x_qual, [0, 0, 5])
qual_md = sk.trimf(x_qual, [0, 5, 10])
```

### Finding the Minimum (Fuzzy AND)

```python
fuzzy_out = sk.fuzzy_and(x_qual, qual_lo, x_qual, qual_md)
```

### Plot

```python
plt.plot(x_qual, qual_lo)
plt.plot(x_qual, qual_md)
plt.plot(fuzzy_out[0], fuzzy_out[1])
plt.grid()
```

# Triangular Norms

```python
#Defininf the T-Norm Function
def t_norm(mfx, mfy):
    tnorm = np.fmin(mfx, mfy)
    return tnorm
```

```python
x = np.arange(0, 100, 1)
#Defining sigmoidal membership function
full_speed = sk.sigmf(x, 80, 2)
slow = sk.sigmf(x, 30, 2)
#Finding the intersection
norm_t = t_norm(full_speed, slow)
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.plot(x, norm_t, 'k', linewidth=1.5, label='Triangular Norms')
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```

# Triangular Co-Norms

```python
def t_cornom(mfx, mfy):
    tnorm = np.fmax(mfx, mfy)
    return tnorm

#Defining sigmoidal membership function
full_speed = sk.sigmf(x, 80, 2)
slow = sk.sigmf(x, 20, 2)

#Finding the Intersection
cornom_t = t_cornom(full_speed, slow)
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.plot(x, cornom_t, 'k', linewidth=1.5, label='Triangular Norms')
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```

```python
def t_conorm(mfx, mfy):
    tnorm = np.fmax(mfx, mfy)
    return tnorm

full_speed = sk.sigmf(x, 50, 0.1)
slow = sk.sigmf(x, 50, 2.0)

cornom_t = t_conorm(full_speed, slow)
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.plot(x, cornom_t, 'k', linewidth=1.5, label='Triangular Norms')
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```
